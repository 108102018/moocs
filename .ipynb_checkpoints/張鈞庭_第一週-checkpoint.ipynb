# 挑戰題一：買相機

```python
%pylab inline
```

```python
k=1000
a=100*k
b=100*k+150*k
c=100*k+150*k+230*k
d=100*k+150*k+230*k+80*k
plus = [a, b, c, d ]
plot(plus)
```

```python
labels = "100k", "150k", "230k", "80k"
size = [100000, 150000, 230000, 80000]
plt.pie(size , labels = labels,autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```
